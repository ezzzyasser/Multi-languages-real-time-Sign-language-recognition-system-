```python
import cv2
import numpy as np
from skimage import color, transform, feature
from joblib import load
import mediapipe as mp

# Load the trained SVM model
model_path = r"E:\gp_final\image\recogntion\svm\version_2\svm_asl_model_2.joblib"
model = load(model_path)

# Function to preprocess the image and extract HOG features
def preprocess_image(image):
    try:
        image = transform.resize(image, (64, 64))  # Resize to 64x64
        image = color.rgb2gray(image)  # Convert to grayscale
        hog_features = feature.hog(image, pixels_per_cell=(8, 8), cells_per_block=(2, 2))
        return hog_features
    except Exception as e:
        print(f"Error in preprocessing image: {e}")
        return None

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False,
                       max_num_hands=1,
                       min_detection_confidence=0.5,
                       min_tracking_confidence=0.5)

# Initialize the webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture frame. Exiting...")
            break

        # Convert the frame to RGB as MediaPipe expects RGB images
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Use MediaPipe to detect hands in the frame
        results = hands.process(frame_rgb)
        
        if results.multi_hand_landmarks:
            # A hand is detected, proceed with preprocessing and prediction
            hog_features = preprocess_image(frame_rgb)
            
            if hog_features is not None:
                print(f"HOG Features: {hog_features[:10]}")  # Print first 10 HOG features for inspection
                # Reshape for the model (1 sample, number of features)
                hog_features = np.array(hog_features).reshape(1, -1)
                print(f"HOG Features Shape: {hog_features.shape}")
                
                # Perform inference
                try:
                    predicted_index = model.predict(hog_features)[0]
                    print(f"Predicted Index: {predicted_index}")
                except Exception as e:
                    print(f"Error during prediction: {e}")
                    predicted_index = '?'
                
                # Draw hand landmarks on the frame
                for hand_landmarks in results.multi_hand_landmarks:
                    mp.solutions.drawing_utils.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                
                # Display the prediction on the frame
                cv2.putText(frame, f'Predicted Index: {predicted_index}', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
            else:
                print("HOG features extraction failed.")
                cv2.putText(frame, 'Error in preprocessing', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
        else:
            cv2.putText(frame, 'No hand detected', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
        
        # Display the resulting frame
        cv2.imshow('Real-Time Sign Language Recognition', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
finally:
    cap.release()
    cv2.destroyAllWindows()
    hands.close()
```
