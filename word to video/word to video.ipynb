```python
import os
import cv2

# Define the path to your dataset
dataset_path = r"E:\gp_final\word to video\videos_by_gloss"

def list_subfolders(dataset_path):
    subfolders = [f for f in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, f))]
    return subfolders

def play_video(video_path):
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print(f"Error: Cannot open video {video_path}")
        return
    
    # Get the frame rate of the video
    fps = cap.get(cv2.CAP_PROP_FPS)
    if fps == 0:
        print("Error: Unable to determine frame rate of the video")
        return
    
    delay = int(1000 / fps)  # Calculate delay in milliseconds

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Display the frame in a popup window
        cv2.imshow('Video', frame)

        # Wait for the appropriate time to match the video's frame rate
        if cv2.waitKey(delay) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

def get_video_path(subfolder_name):
    subfolder_path = os.path.join(dataset_path, subfolder_name)
    if os.path.isdir(subfolder_path):
        video_files = [f for f in os.listdir(subfolder_path) if f.endswith(('.mp4', '.avi', '.mov', '.mkv'))]
        if video_files:
            return os.path.join(subfolder_path, video_files[0])
    return None

# List the subfolder names
subfolders = list_subfolders(dataset_path)
print("You may write one of the following words :")
for subfolder in subfolders:
    print(subfolder)

# Input cell to get the subfolder name and play the video
subfolder_name = input("Enter the name of the subfolder: ").strip()
video_path = get_video_path(subfolder_name)

if video_path:
    play_video(video_path)
else:
    print(f"No video file found in the subfolder: {subfolder_name}")
```
